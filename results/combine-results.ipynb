```python
import pandas as pd
import os
import re

# Function to flatten a CSV to a single row
def flatten_csv(file_path, case_name):
    df = pd.read_csv(file_path)
    flat_dict = {}
    for _, row in df.iterrows():
        source = row['ENERGY SOURCE'].replace(' ', '_')  # Replace spaces with underscores for column names
        for col in df.columns[1:]:
            key = f"{source}_{col}"
            flat_dict[key] = row[col]
    return pd.Series(flat_dict, name=case_name)

# List of CSV file paths (hardcoded for the two files)
# csv_files = [
#     'case-1-power_optimization_results_20250910_131756.csv',
#     'case1-power_optimization_results_20250910_180037.csv'
# ]

csv_files = [f for f in os.listdir('.') if f.endswith('.csv')]

# Initialize an empty list to hold rows
rows = []

# Function to extract case number from filename
def extract_case_number(filename):
    # Use regex to find 'case' followed by a number (allowing for variations like 'case-1' or 'case1')
    match = re.search(r'case-?(\d+)', filename, re.IGNORECASE)
    if match:
        return f'case {match.group(1)}'
    return filename  # Fallback to filename if no case number found

# Process each file
for file_path in csv_files:
    case_name = extract_case_number(os.path.basename(file_path))
    row = flatten_csv(file_path, case_name)
    rows.append(row)

# Combine into a single DataFrame
combined_df = pd.DataFrame(rows)

# Fill NaN with empty string (if any)
combined_df = combined_df.fillna('')

# Save to CSV file
combined_df.to_csv('combined.csv', index=True)

# Optional: Print the CSV content
print(combined_df.to_csv(index=True))
```
